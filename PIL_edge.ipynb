```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
### The purpose of edge detected image is to emphasize features of human postures, and it should be ALWAYS used after
### merging with the origional image. However, experimenting with these edge images will be greatly appreciated!

from PIL import Image, ImageFilter
def image_edge(im):
    copy = (im * 255).astype(np.uint8)
    image = Image.fromarray(copy)
    image = image.convert("L") 
    image = image.filter(ImageFilter.FIND_EDGES) 
 #image_arr is the matplotlib format of images
    image_arr = np.array(image.convert('RGB'))
    image_arr = image_arr.astype(np.float32)/255
    return image_arr
```

```python
# Converts plt.imread() image to pil image
def plt2pil(im):
    im = (im * 255).astype(np.uint8)
    return Image.fromarray(im)
```

```python
#Test image on my local device
im = "Please choose a plt.read() image as im"
plt.imshow(im)
```

```python
#Correspoding edge image
plt.imshow(image_edge(im))
```

```python
#Converts the edge detected image back to pil image for better imshow image quality
plt2pil(image_edge(im))
```
